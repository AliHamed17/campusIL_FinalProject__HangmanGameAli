```python
import random

# Global Constants
MAX_TRIES = 6
num_of_tries = 0
old_letters_guessed = []
secret_word = ""

# Hangman Drawing
HANGMAN_PHOTOS = {
    1: """\
     x-------x
     |
     |
     |
     |
     |""",
    2: """\
     x-------x
     |       |
     |       0
     |
     |
     |""",
    3: """\
     x-------x
     |       |
     |       0
     |       |
     |
     |""",
    4: """\
     x-------x
     |       |
     |       0
     |      /|\\
     |
     |""",
    5: """\
     x-------x
     |       |
     |       0
     |      /|\\
     |      /
     |""",
    6: """\
     x-------x
     |       |
     |       0
     |      /|\\
     |      / \\
     |"""
}

# Function to print the opening screen
def print_opening_screen():
    print("Welcome to the game Hangman")

    # ASCII art for Hangman title
    hangman_art = """\
     _ _
    | | | |
    | |__| | __ _ _ __ __ _ _ __ ___ __ _ _ __
    | __ |/ _' | '_ \\ / _' | '_ ' _ \\ / _' | '_ \\
    | | | | (_| | | | | (_| | | | | | | (_| | | | |
    |_| |_|\__,_|_| |_|\__, |_| |_| |_|\__,_|_| |_|
                         __/ |
                        |___/"""
    print(hangman_art)

    # Display a random number of allowed attempts
    allowed_attempts = random.randint(5, 10)
    print(f"Allowed Attempts: {allowed_attempts}")

# Function to read the secret word from a file based on index
def get_secret_word(file_path, index):
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            words = content.split()

            if 0 <= index < len(words):
                secret_word = words[index].lower()
                return secret_word
            else:
                return None
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return None

# Main Hangman game logic
def hangman(secret_word):
    global num_of_tries
    global old_letters_guessed

    while num_of_tries < MAX_TRIES:
        display_word = show_hidden_word(secret_word, old_letters_guessed)
        print(display_word)

        # Get user input for a letter
        guess = input("Guess a letter: ").lower()

        # Validate the input and handle repeated guesses
        if not guess.isalpha() or len(guess) != 1 or guess in old_letters_guessed:
            print("X")
            print(" -> ".join(sorted(old_letters_guessed)))
            continue

        # Update guessed letters and check if guess is incorrect
        if try_update_letter_guessed(guess, old_letters_guessed):
            if guess not in secret_word:
                num_of_tries += 1
                print(HANGMAN_PHOTOS[num_of_tries])

        # Check for a win and break the loop if the game is won
        if check_win(secret_word, old_letters_guessed):
            print("WIN")
            break

    # Display losing message if all attempts are used
    if num_of_tries == MAX_TRIES:
        print("LOSE")
        print(f"The correct word was: {secret_word}")

# Check if the player has won by guessing all letters
def check_win(secret_word, old_letters_guessed):
    for char in secret_word:
        if char not in old_letters_guessed:
            return False
    return True

# Display the current state of the word with guessed letters
def show_hidden_word(secret_word, old_letters_guessed):
    displayed_word = ""
    for char in secret_word:
        if char in old_letters_guessed:
            displayed_word += char + " "
        else:
            displayed_word += "_ "
    return displayed_word.strip()

# Validate and update the guessed letters
def try_update_letter_guessed(letter_guessed, old_letters_guessed):
    if check_valid_input(letter_guessed, old_letters_guessed):
        old_letters_guessed.append(letter_guessed)
        return True
    else:
        print("X")
        print(" -> ".join(sorted(old_letters_guessed)))
        return False

# Validate the user's input for a guessed letter
def check_valid_input(letter_guessed, old_letters_guessed):
    if len(letter_guessed) != 1 or not letter_guessed.isalpha():
        return False

    letter_guessed = letter_guessed.lower()

    if letter_guessed in old_letters_guessed:
        return False

    return True

# Choose a word from the file based on index and handle circular indexing
def choose_word(file_path, index):
    # Read words from the file
    with open(file_path, 'r') as file:
        words_list = file.read().split()

    # Get the total number of unique words
    unique_words = set(words_list)
    num_unique_words = len(unique_words)

    # Calculate the adjusted index (considering circular fashion)
    adjusted_index = (index - 1) % len(words_list)

    # Get the word at the adjusted index
    secret_word = words_list[adjusted_index]

    return num_unique_words, secret_word


```

```python

def main():
    print_opening_screen()
    file_path = input("Enter the file path: ")
    index = int(input("Enter index: "))
    print("Let's start!\n")
    global secret_word
    secret_word = get_secret_word(file_path, index)

    # Start the Hangman game if the secret word is successfully loaded
    if secret_word is not None:
        hangman(secret_word)

if __name__ == "__main__":
    main()
```
